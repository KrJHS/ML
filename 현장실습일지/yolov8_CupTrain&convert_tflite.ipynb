```python
# yaml 파일을 학습이 가능하도록 경로 설정.
# key-value 데이터인 dict 데이터타입으로 data['train'], data['val'], data['nc'], data['names'] 에 넣어주는데,
# 가장 중요한 부분은 데이터 경로 설정임.

import yaml

data = { 'train' : '/content/drive/MyDrive/eco_cup6/yolo8_dataset/train/images',
         'val' : '/content/drive/MyDrive/eco_cup6/yolo8_dataset/valid/images',
         'test' : '/content/drive/MyDrive/eco_cup6/yolo8_dataset/test/images',
         'names' : ['cup-glass-', 'cup-paper-', 'cup-plastic-', 'straw-paper-', 'straw-plastic-', 'tumbler'],
         'nc' : 6 }

with open('/content/drive/MyDrive/eco_cup6/yolo8_dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)


with open('/content/drive/MyDrive/eco_cup6/yolo8_dataset/data.yaml', 'r') as f:
  cup_yaml = yaml.safe_load(f)
  display(cup_yaml)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # load a pretrained YOLOv8n detection model load a pretrained YOLOv8n detection model YOLOv8n이 가장 가볍고 단순한 모델이다. Detect 방식으로 학습할 예정. 관련 문서:https://docs.ultralytics.com/tasks/detect/#export
                            #Detect, Segment, Classify, Pose 등 다양한 학습 방식이있다.
```

```python
print(type(model.names), len(model.names))

print(model.names) #기존 80개의 데이터 라벨링을 가지고 있다. 하지만 새로 커스텀 하기 때문에 쓰지않는다.
```

```python
model.train(data='/content/drive/MyDrive/eco_cup6/yolo8_dataset/data.yaml', epochs=200, patience=30, batch=32, imgsz=640) #200번 반복, 30번의 반복중 성능 향상이 없으면 중단, 배치 크기 32, 학습 이미지 크기 640x640
```
